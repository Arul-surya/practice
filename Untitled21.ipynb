```python
impression = dfmerge_table[dfmerge_table['Metric']=='Impressions']
```

```python
grouping_year_month = impression.groupby(['month','year'])['Value'].sum()
max_impression_months = grouping_year_month.sort_values(ascending = False).head(3).reset_index()
```

```python
max_impression_months
```

```python
max_impressions = impression.groupby(['Campaign','Model'])['Value'].sum().sort_values(ascending=False).reset_index().head(3)
#final_max_impressions = max_impressions
#.sort_values(ascending=False)#.head(3)
print(max_impressions)
```

```python
Future_Market=impression[impression['Message type']=='Future Market']
```

```python
Future_Market = Future_Market.groupby(['Message type','Campaign'])['Value'].value_counts().reset_index()
#.head(3)#Future Market
Future_Market.sort_values('Value',ascending=False).head(3)
```

```python
Near_Market_table = impression[impression['Message type'] == 'Near Market']
#Near Market table
```

```python
Near_Market = Near_Market_table.groupby(['Message type','Campaign'])['Value'].value_counts().reset_index()
#.head(4)#Near Market
final_Near_Market=Near_Market.sort_values('Value',ascending=False).head(4)
final_Near_Market
```
